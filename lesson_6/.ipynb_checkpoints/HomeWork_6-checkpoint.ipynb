```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import differential_evolution
import time
```

## Уровень 0

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-5, 5), (-5, 5)]
result = differential_evolution(ackley, bounds, seed=42)
```

### 1. Эксперимент с Strategy

```python
str = ['best1bin', 'best1exp', 'rand1exp', 'randtobest1exp', 'currenttobest1exp', 'best2exp', 'rand2exp', 'randtobest1bin', 
       'currenttobest1bin', 'best2bin', 'rand2bin', 'rand1bin']
nit = []
for i in range(len(str)):
    start = time.time()
    result = differential_evolution(ackley, bounds, strategy = str[i], seed=42)
    stop = time.time()
    duration = stop - start
    nit.append(result.nit)
    print('Стратегия:', str[i], '-- Итераций:', result.nit, '-- мин:',
          result.x, '-- ф-я:', result.fun, '-- время:', round(duration, 2))
```

```python
plt.figure(figsize=(60, 20))
plt.bar(str, height=nit)
plt.tick_params(axis='both', which='major', labelsize=26)
plt.show()
```

### 2. Экспериамент с Popsize

```python
nit =[]
for i in range(15):
    start = time.time()
    result = differential_evolution(ackley, bounds, popsize = i, seed=42)
    stop = time.time()
    duration = stop - start
    nit.append(result.nit)
    print('Popsize:', i, '-- Итераций:', result.nit, '-- мин:',
          result.x, '-- ф-я:', result.fun, '-- время:', round(duration, 2))
```

```python
plt.figure(figsize=(60, 20))
plt.bar(range(15), height=nit)
plt.tick_params(axis='both', which='major', labelsize=26)
plt.show()
```

### 3. Экспериамент с Mutation

```python
i = 0.
nit=[]
while i < 2.:
    start = time.time()
    result = differential_evolution(ackley, bounds, mutation = i,seed=42)
    stop = time.time()
    duration = stop - start
    nit.append(result.nit)
    print('Mutation:', round(i, 1), '-- Итераций:', result.nit, '-- мин:',
          result.x, '-- ф-я:', result.fun, '-- время:', round(duration, 2))
    i+=0.1
```

```python
plt.figure(figsize=(60, 20))
plt.bar(range(20), height=nit)
plt.tick_params(axis='both', which='major', labelsize=26)
plt.show()
```

### 4. Эксперимент с Recombination

```python
i = 0.
nit = []
while i < 1:
    start = time.time()
    result = differential_evolution(ackley, bounds, recombination = i,seed=42)
    stop = time.time()
    duration = stop - start
    nit.append(result.nit)
    print('Recombination:', round(i, 1), '-- Итераций:', result.nit, '-- мин:',
          result.x, '-- ф-я:', result.fun, '-- время:', round(duration, 2))
    i+=0.1
```

```python
plt.figure(figsize=(60, 20))
plt.bar(range(11), height=nit)
plt.tick_params(axis='both', which='major', labelsize=26)
plt.show()
```

### 5. Экспериамент с Updating

```python
updating = ['immediate', 'deferred']
nit = []
for i in range(len(updating)):
    start = time.time()
    result = differential_evolution(ackley, bounds, updating = updating[i], seed=42)
    stop = time.time()
    duration = stop - start
    nit.append(result.nit)
    print('Updating:', updating[i], '-- Итераций:', result.nit, '-- мин:',
          result.x, '-- ф-я:', result.fun, '-- время:', round(duration, 2))
```

```python
plt.figure()
plt.bar(updating, height=nit)
plt.tick_params(axis='both', which='major', labelsize=26)
plt.show()
```

## Уровень 1

#### bounds = [(-5, 5), (-5, 5)]

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-5, 5), (-5, 5)]
start = time.time()
result = differential_evolution(ackley, bounds, strategy='best1bin',
                                popsize = 8, mutation=0.3, recombination=1.0, updating='immediate', seed=42)
stop = time.time()
duration = stop-start

print('Итераций:', result.nit, '-- мин:',result.x, '-- ф-я:', result.fun, '-- время:', round(duration, 2))
```

```python
x = [np.linspace(-5, 5, 50), np.linspace(-5, 5, 50)]

plt.figure(figsize=(7,7))
plt.axes(projection='3d')
plt.plot(x[0], x[1], ackley(x))
plt.plot(result.x[0], result.x[1], result.fun, 'ro')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

### Уровень 0.

#### Нормальное распределение

```python
norm_1 = sts.norm(loc=2, scale=0.5)
norm_2 = sts.norm(loc=2, scale=1)
norm_3 = sts.norm(loc=2, scale=2)
norm_4 = sts.norm(loc=1, scale=1)
norm_5 = sts.norm(loc=0, scale=1.5)
```

```python
x = np.linspace(-4, 6, 100)

plt.figure()
plt.plot(x, norm_1.pdf(x))
plt.plot(x, norm_2.pdf(x))
plt.plot(x, norm_3.pdf(x))
plt.plot(x, norm_4.pdf(x))
plt.plot(x, norm_5.pdf(x))
plt.show()
```

#### Экспоненциальное распределение

```python
x = np.linspace(0, 8, 100)

plt.figure()
plt.plot(x, sts.expon(0).pdf(x))
plt.plot(x, sts.expon(0.5).pdf(x))
plt.plot(x, sts.expon(1).pdf(x))
plt.plot(x, sts.expon(1.5).pdf(x))
plt.plot(x, sts.expon(2).pdf(x))
plt.show()
```

#### Распределение Стьюдента

```python
x = np.linspace(-8, 8, 100)

# plt.figure(figsize=(10,7))
plt.plot(x, sts.t(4).pdf(x))
plt.plot(x, sts.t(20).pdf(x))
plt.plot(x, sts.t(0).pdf(x))
plt.plot(x, sts.t(1000).pdf(x))
plt.plot(x, sts.t(-10).pdf(x))
plt.show()
```

```python
from math import log2
```

#### В первом случаем речь идет о разновероятных событиях, поэтому используем формулу Шеннона

```python
def shennon(a,b,c):
    summ = a+b+c
    return a/summ*log2(1/(a/summ)) + b/summ*log2(1/(b/summ)) + c/summ*log2(1/(c/summ))
result = shennon(54, 4, 28)
print("Количество информации о цвете светофора в текущий момент:", round(result, 4))
```

#### Во втором случае речь идет о равновероятных событий, поэтому используем формулу Хартли

```python
result = log2(1/(28/84))
print("Количество информации о цвете светофора в текущий момент:", round(result, 4))
```

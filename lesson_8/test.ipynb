```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

#### Нормальное распределение

```python
mu = 2
sigma = 0.5

norm = sts.norm(loc=mu, scale=sigma)
norm.rvs(20)
```

#### Функция кумулятивной плотности нормального распределения (Показывает совокупную вероятность на оси Y)

```python
x = np.linspace(0, 4, 100)

cdf = norm.cdf(x)
plt.figure()
plt.plot(x, cdf)
plt.show()
```

#### Функция плотности вероятности нормального распределения (Вероятность значений)

```python
x = np.linspace(0, 4, 100)

pdf = norm.pdf(x)
plt.figure()
plt.plot(x, pdf)
plt.show()
```
